```julia
# question 1

u1(X1, X2) = 0.2*log(1 + X1 + X2)
u2(X1, X2) = 0.2*(exp(X1))

X1_range = 0.1:0.1:1
X2_range = 0.1:0.1:1

x_displaced = [X1 + u1(X1,X2) for X1 in X1_range, X2 in X2_range]
y_displaced = [X2 + u2(X1,X2) for X1 in X1_range, X2 in X2_range]

plt = plot(x_displaced, y_displaced, legend=false, xlabel="X", ylabel = "X", title = "displaced shape of body")

x_displaced_2 = [X1 + u1(X1,X2) for X2 in X2_range, X1 in X1_range]
y_displaced_2 = [X2 + u2(X1,X2) for X2 in X2_range, X1 in X1_range]

plot!(plt, x_displaced_2, y_displaced_2, label="displaced")

original_x = [0.1, 1.0, 1.0, 0.1, 0.1]
original_y = [0.1, 0.1, 1.0, 1.0, 0.1]
plot!(plt, original_x, original_y, color=:black, linewidth=2, linestyle=:dash, label="Original")
```

Using these arrays, it plots families of lines: one set obtained by varying X1 for fixed X2 (approximately horizontal characteristic lines) and another set obtained by varying X2 for fixed X1 (approximately vertical characteristic lines).
